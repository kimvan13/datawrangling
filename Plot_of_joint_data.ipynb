```r
library(tidyverse)
library(ggplot2)
```

```r
final <- "Joined_df.csv" %>%
    read_csv()
```

```r
head(final)
```

```r
par(mfrow=c(1, 3))
boxplot(final$Student_enrolled_primary, data = final, 
        main = "Student enrolled primary",
        notch = F, 
        varwidth = TRUE, 
        col = c("green")
)
boxplot(final$LabourForce , data = final, 
        main = "Labour Force ",
        notch = F, 
        varwidth = TRUE, 
        col = c("green")
)
boxplot(final$TotalUnemployment , data = final, 
        main = "Total Unemployment ",
        notch = F, 
        varwidth = TRUE, 
        col = c("green")
)
```

The first plot shows that the median of students enrolled in primary level is 360, the minimal is under 350 and maximal is over 380. The third plot shows that the median of labour force is between 2,400,000 and 2,500,000, the minimal is under 2200000 and maximal is roughly 2700000. The second plot shows that the median of total unemployment is between 5.0 and 5.5, the minimal is under 4.0 and maximal is over 6.5. All boxplots does not have any outlier.

```r
ggplot(data = final, aes(x = Year, y = TotalUnemployment)) +
  geom_point() + geom_line(linetype = "dashed")+
  labs(x = "Year",
       y = "Total unemployment",
       title = "Total unemployment in New Zealand")
```

#### From the time series plot we can see that the unemployment rate in New Zealand rises sharply from 2005 to 2010, and goes down until 2017.

```r
ggplot(data = final, aes(x = Year, y = LabourForce)) +
geom_line(linetype = "dashed")+
  geom_point() +
  labs(x = "Year",
       y = "Labour force",
       title = "Labour force in New Zealand")
```

#### The labour force in New Zealand increases from 2005 to 2017 sharper and sharper.

```r
ggplot(data = final, aes(x = Year, y = Student_enrolled_primary)) +
  geom_point() + geom_line(linetype = "dashed")+
  labs(x = "Year",
       y = "Student enrolled primary",
       title = "Student enrolled primary in New Zealand")
```

#### The student enrollment in primary in New Zealand goes down from 2005 to 2010, then increases rapidly.

```r
ggplot(data = final, aes(x = Year, y = Government_Expenditure, group=1)) +
geom_line(linetype = "dashed")+
  geom_point() + 
  labs(x = "Year",
       y = "Government expenditure",
       title = "Government expenditure in New Zealand")
```

#### The Government expenditure in primary in New Zealand goes down sharply from 2005 to 2010, then increases rapidly
